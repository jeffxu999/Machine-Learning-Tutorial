```
# Multiple Linear Regression - Machine Learning Tutorial

########################################################

import pandas as pd
from sklearn import datasets, linear_model, metrics
import matplotlib.pyplot as plt

########################################################

diabetes_data = datasets.load_diabetes()

variables = pd.DataFrame(diabetes_data.data)
diseases = pd.DataFrame(diabetes_data.target)
labels = pd.DataFrame(diabetes_data.feature_names)
descriptions = diabetes_data.DESCR

training_variables = variables[0:350]
test_variables = variables[350:]

training_diseases = diseases[0:350]
test_diseases = diseases[350:]

########################################################

our_model = linear_model.LinearRegression()

our_model.fit(training_variables,training_diseases)

########################################################

predicted_diseases = our_model.predict(test_variables)

error = metrics.mean_squared_error(test_diseases,predicted_diseases)

r_squared = metrics.r2_score(test_diseases,predicted_diseases)

coefficients = our_model.coef_

plt.scatter(training_variables[2],training_diseases)

plt.show()
```
