```
# Hierarchical Clustering - Machine Learning Tutorial

###############################################################

from sklearn import datasets, cluster
import matplotlib.pyplot as plt

###############################################################

blob_values, blob_labels = datasets.make_blobs(n_samples=300,centers=3)

our_hierarchical_cluster = cluster.AgglomerativeClustering(n_clusters=3)

###############################################################

our_hierarchical_cluster.fit(blob_values)

clusters = our_hierarchical_cluster.labels_

###############################################################

print("Actual: Orange = 0, Green = 1, & Blue = 2")

for index in range(len(blob_values)):
    if blob_labels[index]==0:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="orange")
    elif blob_labels[index]==1:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="green")
    else:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="blue")
        
plt.show()

print("Predicted: Orange = 0, Green = 1, & Blue = 2")

for index in range(len(blob_values)):
    if clusters[index]==0:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="purple")
    elif clusters[index]==1:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="red")
    else:
        plt.scatter(blob_values[index][0],blob_values[index][1],c="black")
        
plt.show()
```
