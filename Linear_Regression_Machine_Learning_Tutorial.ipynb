```
# Linear Regression - Machine Learning Tutorial

#########################################################

import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import random

#########################################################

x = [i for i in range(100)]

y = [2*random.randint(i-25,i+25)+5 for i in range(100)]

x_train = x[0:80]
x_test = x[80:]

y_train = y[0:80]
y_test = y[80:]

#########################################################

slope, intercept = np.polyfit(x_train,y_train,1)

#########################################################

y_predicted = []

for x_value in x_test:
    prediction = x_value*slope+intercept
    y_predicted.append(prediction)

r_squared = metrics.r2_score(y_test,y_predicted)

error = metrics.mean_squared_error(y_test,y_predicted)

#########################################################

y_predicted_graph = []

for x_value in x:
    prediction = x_value*slope+intercept
    y_predicted_graph.append(prediction)

plt.scatter(x,y,c="red",marker="v")

plt.plot(x,y_predicted_graph)

plt.show()
```
