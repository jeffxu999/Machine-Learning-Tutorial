```
# Quadratic Regression - Machine Learning Tutorial

#######################################################

from sklearn import datasets, metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#######################################################

housing_data = datasets.load_boston()

features = pd.DataFrame(housing_data.data, columns=housing_data.feature_names)

targets = pd.DataFrame(housing_data.target)

feature_selected = "LSTAT"

training_data = features[feature_selected][0:404]
test_data = features[feature_selected][404:]

training_targets = targets[0:404]
test_targets = targets[404:]

#######################################################

a,b,c = np.polyfit(training_data,training_targets,2)

#######################################################

predicted_training = []

for x in training_data:
    predicted = a*(x*x) + b*x + c
    predicted_training.append(predicted)
    
print("Prediction Data")
plt.scatter(training_data,predicted_training,c="red")
plt.scatter(training_data,training_targets)
plt.show()

predicted_test = []

for x in test_data:
    predicted = a*(x*x) + b*x + c
    predicted_test.append(predicted)
    
print("Test Data")
plt.scatter(test_data,predicted_test,c="orange")
plt.scatter(test_data,test_targets)
plt.show()

accuracy = metrics.r2_score(test_targets,predicted_test)

print(accuracy)
```
