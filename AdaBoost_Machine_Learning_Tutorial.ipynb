```
# AdaBoost - Machine Learning Tutorial

#######################################################

from sklearn import datasets, utils, ensemble
import pandas as pd

#######################################################

wine_data = datasets.load_wine()

inputs, outputs = utils.shuffle(wine_data.data,wine_data.target,random_state=111)

features = pd.DataFrame(inputs,columns=wine_data.feature_names)

targets = pd.DataFrame(outputs)

training_data = features[0:142]
test_data = features[142:]

training_target = targets[0:142]
test_target = targets[142:]

#######################################################

our_forest = ensemble.RandomForestClassifier(max_depth=1)

our_forest.fit(training_data,training_target)

weak_accuracy_test = our_forest.score(test_data,test_target)

print("Weak Test:",weak_accuracy_test)

#######################################################

our_adaboost = ensemble.AdaBoostClassifier(ensemble.RandomForestClassifier(max_depth=1))

our_adaboost.fit(training_data,training_target)

strong_accuracy_test = our_adaboost.score(test_data,test_target)

print("Strong Test:",strong_accuracy_test)

#######################################################

strong_predictions = pd.DataFrame(our_adaboost.predict(test_data))

weak_predictions = pd.DataFrame(our_forest.predict(test_data))

print("Strong:", strong_predictions.head())
print("Weak:", weak_predictions.head())
print("Target:", test_target.head())
```
