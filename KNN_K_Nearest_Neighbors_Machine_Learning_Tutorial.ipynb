```
# KNN K Nearest Neighbors - Machine Learning Tutorial

################################################################

from sklearn import datasets, neighbors
import pandas as pd
import matplotlib.pyplot as plt

################################################################

iris_data = datasets.load_iris()

iris_features = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

pedal_width = iris_features["petal width (cm)"]

iris_types = pd.DataFrame(iris_data.target)

train_variables = pedal_width[0:120]
test_variables = pedal_width[120:]

train_types = iris_types[0:120]
test_types = iris_types[120:]

################################################################

our_model = neighbors.KNeighborsClassifier()

train_variables_vector = train_variables.reshape(-1,1)

our_model.fit(train_variables_vector, train_types)

################################################################
              
test_variables_vector = test_variables.reshape(-1,1)

mean_accuracy = our_model.score(test_variables_vector, test_types)

print(mean_accuracy)

prediction1 = our_model.predict([[0.3]])
prediction2 = our_model.predict([[1.4]])
prediction3 = our_model.predict([[2]])

graph_figure = plt.figure(figsize=(10,20))

ax1 = graph_figure.add_subplot(3,1,1)
ax2 = graph_figure.add_subplot(3,1,2)
ax3 = graph_figure.add_subplot(3,1,3)

ax1.scatter(pedal_width, iris_types)
ax1.scatter(0.3, prediction1, c="red", linewidths="10")
ax1.set_title("Input: 0.3 | Output: "+str(prediction1))

ax2.scatter(pedal_width, iris_types)
ax2.scatter(1.4, prediction2, c="orange", linewidths="10")
ax2.set_title("Input: 1.4 | Output: "+str(prediction2))

ax3.scatter(pedal_width, iris_types)
ax3.scatter(2, prediction3, c="purple", linewidths="10")
ax3.set_title("Input: 2 | Output: "+str(prediction3))

plt.show()
```
