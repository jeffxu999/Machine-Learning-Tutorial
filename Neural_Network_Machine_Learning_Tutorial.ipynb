```
# Neural Network - Machine Learning Tutorial

##########################################################

from sklearn import datasets, neural_network
import matplotlib.pyplot as plt
import numpy as np

##########################################################

digits_data = datasets.load_digits()

images = digits_data.images

images_vector = digits_data.data

targets = digits_data.target

training_images = images_vector[0:1437]
test_images = images_vector[1437:]

training_targets = targets[0:1437]
test_targets = targets[1437:]

##########################################################

our_neural_net = neural_network.MLPClassifier()

our_neural_net.fit(training_images,training_targets)

##########################################################

accuracy = our_neural_net.score(test_images, test_targets)

predictions = our_neural_net.predict(test_images)

for index in range(100):
    print("Predicted : ", our_neural_net.predict(np.reshape(test_images[index],(1,-1))))
    print("Actual : ",test_targets[index])
    plt.imshow(np.reshape(test_images[index],(8,8)))
    plt.show()
```
