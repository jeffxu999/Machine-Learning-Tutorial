```
# Decision Tree Classifier - Machine Learning Tutorial

###################################################

from sklearn import datasets, tree
import pandas as pd
import numpy as np

###################################################

cancer_data = datasets.load_breast_cancer()

features = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

targets = pd.DataFrame(cancer_data.target)

training_features = features[0:455]
test_features = features[455:]

training_targets = targets[0:455]
test_targets = targets[455:]

###################################################

our_tree = tree.DecisionTreeClassifier()

our_tree.fit(training_features,training_targets)

###################################################

mean_accuracy = our_tree.score(test_features,test_targets)

test_point = test_features.iloc[0]

test_point_vector = test_point.reshape(1,-1)

prediction = our_tree.predict(test_point_vector)

our_predictions = []

for index in range(len(test_features)):
    feature = test_features.iloc[index]
    feature_vector = feature.reshape(1,-1)
    our_prediction = our_tree.predict(feature_vector)
    our_predictions.append(int(our_prediction))
    
our_predictions = pd.DataFrame(our_predictions)

print(our_predictions.head())

print(test_targets.head())
```
